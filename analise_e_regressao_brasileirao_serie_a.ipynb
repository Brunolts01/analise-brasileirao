## 3.0 Criando Super Tabela de Pontuação

Iremos entender agora quais foram os melhores times nos últimos 9 anos, de 2014 a 2023.
Para isso, iremos analisar as colunas de pontos, vitórias, saldo de gols e aproveitamento. Iremos fazer a média dos últimos anos, criar uma nova colunas de pontuação e por fim achar os quartis desses dados.

```python
## Criando nova tabela com média geral de pontos, vitórias, salgo de gols e aproveitamento e por fim criando uma coluna com a soma de todos esses itens.

# Nova tabela
df_super_tabela = df.groupby(['TIME'], as_index=False).agg({'PTS': 'mean', 'V': 'mean', 'SG':'mean', 'APROVEITAMENTO':'mean'}).rename(columns={'PTS': 'MEDIA_PONTOS',
                                                                                                                                               'V':'MEDIA_VITORIAS',
                                                                                                                                               'SG': 'MEDIA_SG',
                                                                                                                                               'APROVEITAMENTO':'MEDIA_APROVEITAMENTO'})
# Nova Coluna
df_super_tabela['PONTUACAO_FINAL'] = df_super_tabela['MEDIA_PONTOS'] + df_super_tabela['MEDIA_VITORIAS'] + df_super_tabela['MEDIA_SG'] + df_super_tabela['MEDIA_APROVEITAMENTO']
```

Tabela pronta iremos ordenar a mesma para ver os melhores times.

```python
# Ordenando tabela para ver os melhores times

df_super_tabela.sort_values(['PONTUACAO_FINAL'], ascending=False)
```

Agora iremos achar os quartis da nossa base. Assim poderemos ter uma visão melhor além da pontuação para entender como os times se dividem.

```python
# Achando quartis
quartis = df_super_tabela['PONTUACAO_FINAL'].quantile([0.25, 0.50, 0.75, 1.0])
q1, q2, q3, q4 = quartis.values
```

```python
# Calculando bins e criando labels para coluna
bins = [df_super_tabela['PONTUACAO_FINAL'].min()-1, q1, q2, q3, q4]
labels = ['Q1', 'Q2', 'Q3', 'Q4']
```

```python
# Criando coluna de quartial
df_super_tabela['QUARTIL'] = pd.cut(df_super_tabela['PONTUACAO_FINAL'], bins=bins, labels=labels)
```

```python
# Tabela final com quartis aplicados
df_super_tabela.sort_values('PONTUACAO_FINAL', ascending=False)
```

```python
# Agrupamento por quartil
df_super_tabela.groupby("QUARTIL").TIME.count()
```

### Conclusões da análise



*   O time com a **maior média de pontos** e **melhor saldo de gols** é o **Palmeiras**.
*   O time que mais levou **cartões amarelos** foi o **Criciúma** e o que menos levou foi o **Corinthians**.
*   O time que mais levou **cartões vermelhos** foi o **Paraná** e o que menos levou foi o **Corinthians**.

## Correlação de Variáveis e Regressão Linear

Nessa segunda parte iremos entender se há váriaveis que se correlacionam e dado o entendimento trabalharemos em um modelo de regressão linear para previsão de valores.

### 1.0 Entendendo correlação das variáveis

Vamos plotar um heatmap para começar a entender se existe relação entre nossas variáveis.

```python
# --> Mapa de correlação

# Deixando apenas colunas necessárias para análise
df_filtrado_2 = df[['POSIÇÃO', 'PTS', 'V', 'E', 'D', 'GP', 'GC', 'SG', 'CA', 'CV']]

# Criando corrrelação
df_filtrado_2 = df_filtrado_2.corr()

# Plotando gráfico
plt.figure(figsize=(20,7))
sns.heatmap(data = df_filtrado_2, square=True, vmin=-1, vmax=1, annot=True)
plt.title("Correlação entre Variáveis")

plt.show();
```

Temos algumas correlações bem fortes no dataset, por exemplo, vitória e pontos tem uma correlação de **0.98**. Isso acontece porque os pontos do brasileirão estão ligado diretamente com as vitórias, cada uma conta 3 pontos na tabela.

Pois bem, mas temos algumas outras que não são tão fortes assim, como **Saldo de Gols e Pontos** que fica em 0.77.

Vamos plotar um gráfico de dispersão que irá nos ajudar a entender melhor essas duas variáveis.

```python
# Plot de gráfico com relação entre variáveis

X = df_filtrado['SG']
Y = df_filtrado['PTS']

plt.figure(figsize=(10,5))
sns.scatterplot(x = X, y = Y)
plt.title("Relação entre Saldo de Gols e Pontos")
plt.xlabel("Saldo de Gols")
plt.ylabel("Pontos")

plt.show();
```

Temos uma relação clara entre essas duas variáveis, quanto maior saldo de gols maior o número de pontos.

Dado o entendimento iremos seguir com nosso modelo de Regressão Linear a fim de **prever os Pontos** **com base no Saldo de Gols**.

Iremos escolher e instanciar nosso modelo, logo depois dividir os dados entre treino e teste, fazer o fit e por fim as previsões nos dados de teste.

```python
# 1. escolher e importar um modelo
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# 2. Instanciar e escolher os hyperparameters
model = LinearRegression()

# 3. Separar os dados entre feature matrix e target vector
X = np.array(df_filtrado['SG']).reshape(-1, 1)
y = np.array(df_filtrado['PTS']).reshape(-1, 1)

# 3.1 Dividir o dataset entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)

# 4. Fit do modelo (treinar)
model.fit(X_train, y_train)

# 5. Fazer previsões em cima de novos dados
y_pred = model.predict(X_test)
```

Para fins de curiosidade, além das métricas prontas so Scikit Learn também seria possível calcular a MAE e MSE com Python puro. Vamos criar as funções e depois calcular nossos erros.

```python
# Funções com Python Puro
# MAE
def MAE(y_real, y_pred):
  return np.sum(np.abs(y_real - y_pred))/len(y_real)

# MSE
def MSE(y_real, y_pred):
  return np.sum((y_real - y_pred)**2/len(y_real))
```

```python
# Retorno das previsões
print(f"MAE com Sklearn: {mean_absolute_error(y_test, y_pred)}")
print(f"MAE com Python: {MAE(y_test, y_pred)}")

print(f"MSE com Sklearn: {mean_squared_error(y_test, y_pred)}")
print(f"MSE com Python: {MSE(y_test, y_pred)}")
```

Calculadas as métricas é possível observar que não obtivemos erros muito altos.

Vamos analisar de maneira visual num gráfico de dispersão traçando uma linha de valores previstos como nosso modelo previu os valores.

```python
# Plotando gráfico de previsões
plt.figure(figsize=(10, 5))
sns.scatterplot(x = X_test.reshape(1, -1)[0], y = y_test.reshape(1, -1)[0])
sns.lineplot(x = X_test.reshape(1, -1)[0], y = y_pred.reshape(1, -1)[0], color='red')
plt.xlabel("Saldo de Gols")
plt.ylabel("Pontos")
plt.title("Previsão de Pontos com base no Saldo de Gols")

plt.show();
```

Vamos fazer a previsão de um valor único agora. Conforme vemos acima, numa média de 30 gols nosso modelo deverá prever um valor de 70 pontos aproximadamente, vamos validar.

```python
# Previsão
v_pred = np.array(30).reshape(1, -1)
model.predict(v_pred)
```
